## 전체 데이터 하나의 데이터 프레임에 저장하기

```python
tl = pd.read_csv('./tag_original.csv', encoding='utf8')
tl = tl.sort_values('tag_list', ascending=True)
tl.rename(columns = {'tag_list': 'Name'}, inplace=True)

instagram = pd.read_csv('./instagram crawling.csv', encoding='utf8')
instagram_df = pd.merge(instagram, tl, on = 'Name', how = 'outer')
instagram_df.drop(axis=1, columns = 'Unnamed: 0', inplace=True)
instagram_df['Count'].fillna(0, inplace=True)
instagram_df['Count'] = instagram_df['Count'].astype('int')

naver = pd.read_csv('./toy project/naver_list.csv', encoding='utf8')
naver.rename(columns = {'tag_list': 'Name'}, inplace=True)
naver_df = pd.merge(naver, tl, on = 'Name', how = 'outer')

fb = pd.read_csv('./toy project/facebook_crawl_data.csv', encoding='utf8')
fb.rename(columns = {'Unnamed: 0':'Name'}, inplace=True)
fb = fb[['Name', 'Search']]
fb_df = pd.merge(fb, tl, on = 'Name', how = 'outer')

tot_df = pd.concat([instagram_df[['Name','Count']], naver_df['naver_like'], fb_df['Search'], tl['original']], axis =1)
tot_df.rename(columns = {"Count":"insta_total_post","Search": "fb_total_search"}, inplace=True)
tot_df
```

## 성별 데이터 변수에 저장하고 class 컬럼 생성하기

```python
gender = pd.read_excel('./toy project/sex_kakao.xlsx',engine ='openpyxl', usecols = 'A:C')
gender.rename(columns={'성별':'Name'}, inplace=True)
gender['Name'].fillna('진격의거인TheFinalSeason', inplace=True)
gender['여성'].fillna(0, inplace=True)
gender['남성'].fillna(0, inplace=True)
gender[['여성', '남성']] = gender[['여성', '남성']].astype('int')

gender_df = pd.merge(gender, tl, on = 'Name', how = 'outer')
gender_df['class'] = None
gender_df['class'][gender_df['여성'] > gender_df['남성']] = 'h'
gender_df['class'][gender_df['여성'] == gender_df['남성']] = 's'
gender_df['class'][gender_df['여성'] < gender_df['남성']] = 'l'
gender_df
```

```python
gender_df1 = pd.merge(tot_df, gender_df[['Name','class']], on = 'Name', how = 'outer')
gender_df1
```

```python
age = pd.read_excel('./toy project/age_kakao.xlsx',engine ='openpyxl', usecols = 'A:G', sheet_name = 'Sheet1')
age.rename(columns={'연령':'Name'}, inplace=True)
age['Name'].fillna('진격의거인TheFinalSeason', inplace=True)
for i in range(1,7):
    if i <= 5:
        age[f'{10*i}대'].fillna(0, inplace=True)
        age[f'{10*i}대'] = age[f'{10*i}대'].astype('int')
    elif i == 6:
        age[f'{10*i}대 이상'].fillna(0, inplace=True)
        age[f'{10*i}대 이상'] = age[f'{10*i}대 이상'].astype('int')
age 
```

```python
age_contents = pd.merge(age[["Name", '40대']], tot_df, on = 'Name', how = 'outer')
age_contents
```

```python
age_sns = round(age_contents.groupby("original").mean(),0)
age_sns
```
